```python


zoom_x = 2.0  # horizontal zoom
zoom_y = 2.0  # vertical zoom
mat = fitz.Matrix(zoom_x, zoom_y)
pdf_path = r"1page.pdf"

base_name = os.path.basename(pdf_path)


all_annotations = {}
with fitz.open(pdf_path) as doc:
    for page in doc:
        this_anns = page.get_text("blocks", sort=True)
        all_annotations[page.number] = []
        for ann in this_anns:
            if ann[6] != 0:
                continue
            x0, y0, x1, y1 = ann[:4]
            text = ann[4]
            block_num = ann[5]
            this_ann = {
                "bbox": [x0 * zoom_x, y0 * zoom_y, x1 * zoom_x, y1 * zoom_y],
                "text": text.strip(),
                "block_num": block_num
            }
            all_annotations[page.number].append(this_ann)
        pix = page.get_pixmap(matrix=mat)  # render page to an image
        pix.save(f"{base_name}-page-{page.number}.png")  # store image as a PNG

total_page = len(all_annotations)
for page_number in range(total_page):
    print(f"page {page_number} has {len(all_annotations[page_number])} annotations")
    img_path = f"{base_name}-page-{page_number}.png"
    output_path = f"{base_name}-output-{page_number}.png"
    annotation = all_annotations[page_number]

    # 可视化
    img = read_image(img_path)
    boxes = [_["bbox"] for _ in annotation]
    labels = [str(_["block_num"]) for _ in annotation]
    boxes = torch.tensor(boxes, dtype=torch.int)

    img_vis = draw_bounding_boxes(img, boxes, width=2, fill=True, labels=labels)
    img_vis = torchvision.transforms.ToPILImage()(img_vis)

    # display output
    img_vis.save(output_path)

    plt.figure(figsize=(15,10))
    # 显示原始图片和打标的图片
    img = mpimg.imread(img_path)
    plt.imshow(img)

    img_anno = mpimg.imread(output_path)
    plt.imshow(img_anno)

```
